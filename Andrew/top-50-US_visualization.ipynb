Going to set up correlation matrix, need to scale the data

```python
# need to drop remaining object columns
# drop title and Artist
spotify_df = spotify_df.drop(columns = ['Title', 'Artist'])
spotify_df
```

```python
# need to convert columns to float type
number_columns = ['Artist_followers', 'Track_number', 'Tracks_in_album', 'danceability', 'energy', 'key', 'loudness' , 'mode', 'speechiness', 'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo', 'duration_ms', 'time_signature']

for column in number_columns:
    spotify_df[column] = spotify_df[column].astype(float)
```

```python
# Create features and target sets
spotify_features = spotify_df.drop(columns = ['Top50_dummy'])
spotify_target = spotify_df['Top50_dummy'] 
```

```python
from sklearn.preprocessing import StandardScaler

# create the scaler instance
scaler = StandardScaler()

# scaling the feature data
spotify_features_scaled_df = scaler.fit_transform(spotify_features)
# making the scaled data into a dataframe
spotify_features_scaled_df = pd.DataFrame(spotify_features_scaled_df)
# setting up the column names
spotify_features_scaled_df.columns = spotify_features.columns
# viewing first 5 rows of dataframe
spotify_features_scaled_df.head()
```

```python
#getting rid of unnamed column
spotify_features_df = spotify_features_scaled_df.iloc[:, 1:]
spotify_features_df = spotify_features_df.drop(columns=['bolero', 'opm'])

spotify_features_df
```

Making a correlation matrix of the scaled trained data

```python
# setting up correlation matrix
spotify_correlation_matrix = spotify_features_df.corr()

spotify_correlation_matrix
```

```python
spotify_correlation_matrix.hvplot.heatmap(height=1000, width=800, colorbar=True, rot=90, title='Spotify Features Correlation Matrix')
```
