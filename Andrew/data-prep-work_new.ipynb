```python
# setting up imports
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
# importing the original csv file with spotify data
spotify_df = pd.read_csv(Path("../Resources/spotify_data.csv"))

# checking csv data with .head function
spotify_df.head(10)
```

First analyzing the data set, getting an idea of size, null values and other parameters

```python
# using df info to check the columns and the contents
spotify_df.info(verbose=True, show_counts=True)
```

```python
# checking to see how many null values are in these columns
spotify_df.isnull().sum()
```

```python
# since the dataset is quite large with a large number of columns, set up loop to check all the columns to see if there are null values
# made this into a function so can call later
def dfNullValCount(data):
    columns_with_null_values = 0
    for column in data.columns:
        if data[column].isnull().sum() != 0:
            columns_with_null_values += 1
            print(f'{column} has null values')
    return columns_with_null_values

null_count = dfNullValCount(spotify_df)
display(null_count)
```

There are 57 columns with null values, ie the initial analysis with just checking all for null wasn't sufficient. For this project we will drop the null values as instructed in the modules, but this may skew the end results

```python
# dropping null values in dataframe
spotify_df = spotify_df.dropna().copy()

#checking to see that all null values are gone
null_count = dfNullValCount(spotify_df)
display(null_count)
```
