```python
# setting up imports
import pandas as pd
import numpy as np
from pathlib import Path
import hvplot.pandas
```

```python
# importing the resampled and trimmed csv file with spotify data
spotify_df = pd.read_csv(Path("../../Resources/spotify_resampled_df.csv"))

# checking csv data with .head function
spotify_df.head(10)
```

studying the different features from the resampled data set vs the top 10 column of interest

collecting the features and target column from the dataframe

```python
# selecting rows from popularity to the end (chopped off target column)
spotify_data_df = spotify_df.iloc[:, 5:]

# displaying to check the data
spotify_data_df.head()
```

checking to see that the data/plots haven't changed

```python
spotify_data_df.hvplot.box(rot=90, title="final feature data", ylabel="Scaled Units", xlabel="Features", height=400, width=1000)
```

setting up a correlation matrix to see if there are any positive correlations between different features of interest

```python
spotify_features_data_df = spotify_data_df.iloc[:, 0:-1]

spotify_features_data_df
```

```python
# setting up correlation matrix
spotify_correlation_matrix = spotify_features_data_df.corr()

spotify_correlation_matrix
```

```python
spotify_correlation_matrix.hvplot.heatmap(height=1000, width=800, colorbar=True, rot=90, title='Spotify Features Correlation Matrix',
                                          cmap=["#75968f", "#a5bab7", "#dfccce", "#ddb7b1", "#550b1d"],)
```
