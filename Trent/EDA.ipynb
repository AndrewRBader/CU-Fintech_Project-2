```python
import pandas as pd
from pathlib import Path

pd.set_option('display.max_columns',200)
```

```python
df_us=pd.read_csv(Path('../Resources/spotify_us_df.csv'))
df_us.head()
```

```python
df_us = df_us.drop(columns= ['Popularity', 'Title', 'Artist', 'Album/Single', 'Genre', 'Genre_new', 'Explicit', 'Album', 'Release_date', 'Cluster','Top10_dummy','Popu_max'])
df_us
```

```python
scaler_variables=['Artist_followers','Track_number','Tracks_in_album','key','loudness','tempo','duration_ms','time_signature']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Create features (X) and target (y) sets
X = df_us.drop(columns = ['Top50_dummy'])
y = df_us['Top50_dummy']
```

```python
# Separate the data
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
scaler=StandardScaler()
```

```python
scaler=scaler.fit(X_train[scaler_variables])
X_train_scaled=scaler.transform(X_train[scaler_variables])
X_test_scaled=scaler.transform(X_test[scaler_variables])
```

```python
model=RandomForestClassifier()
model=model.fit(X_train_scaled,y_train)
pred=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model=GradientBoostingClassifier()
model=model.fit(X_train_scaled,y_train)
pred=model.predict(X_test_scaled)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model=model.fit(X_train_scaled,y_train)
pred=model.predict(X_test_scaled)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier()
model=model.fit(X_train_scaled,y_train)
pred=model.predict(X_test_scaled)
```

```python
print(classification_report(y_test,pred))
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
knn=KNeighborsClassifier()
params_knn={'n_neighbors':np.arange(1,25)}
knn_gs=GridSearchCV(knn,params_knn,cv=5)
knn_gs.fit(X_train_scaled,y_train)
```

```python
knn_best = knn_gs.best_estimator_
print(knn_gs.best_params_)
```

```python
rf=RandomForestClassifier()
params_rf={'n_estimators':[50,100,200]}
rf_gs = GridSearchCV(rf,params_rf, cv=5)
rf_gs.fit(X_train_scaled,y_train)
```

```python
rf_best=rf_gs.best_estimator_
print(rf_gs.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(X_train_scaled,y_train)
```

```python
print('knn: {}'.format(knn_best.score(X_test_scaled, y_test)))
print('rf: {}'.format(rf_best.score(X_test_scaled, y_test)))
print('log_reg: {}'.format(log_reg.score(X_test_scaled, y_test)))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
estimators=[('knn',knn_best),('rf',rf_best),('log_reg',log_reg)]
```

```python
ensemble=VotingClassifier(estimators,voting='hard')
```

```python
ensemble.fit(X_train_scaled,y_train)
ensemble.score(X_test_scaled,y_test)
```

```python
pred=ensemble.predict(X_test_scaled)
print(classification_report(y_test,pred))
```
