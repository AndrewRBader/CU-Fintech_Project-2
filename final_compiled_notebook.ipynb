```python
# Import required libraries 
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.preprocessing import StandardScaler
import hvplot.pandas

# Display all columns 
pd.set_option('display.max_columns', 200)
```

## Data Prep and Cleaning

```python
# Open the CSV file in a Pandas DataFrame
spotify_df = pd.read_csv(Path("Resources/spotify_data.csv"))

# checking spotify dataframe
spotify_df
```

```python
# Drop unnecessary/irrelevant/redundant columns
spotify_df = spotify_df.drop(
    columns=['Album/Single', 'Genre', 'Genre_new', 'Explicit', 'Album', 'Release_date', 'Cluster', 
             'Popu_max', 'Popularity', 'Cluster', 'bing_norm_negative', 'bing_norm_neutral', 
             'bing_norm_positive','Argentina','Australia','Austria','Belgium','Brazil','Canada',
             'Chile','Colombia','Costa Rica','Denmark','Ecuador','Finland','France','Germany',
             'Global','Indonesia','Ireland','Italy','Malaysia','Mexico','Netherlands','New Zealand',
             'Norway','Peru','Philippines','Poland','Portugal','Singapore','Spain','Sweden',
             'Switzerland','Turkey','UK','USA', 'opm', 'bolero', 'jazz', 'reggae'])
spotify_df.head()
```

```python
# seeing number of rows and columns remaining
spotify_df.shape
```

```python
# drop the Uri from the dataset
spotify_df = spotify_df.drop(columns=['Uri'])

# focusing on US data, selecting US data from original dataframe
spotify_us_df = spotify_df[spotify_df['Country'] == "USA"].copy()

# dropping null columns in dataframe - these columns were created by the creators of this data using thier NLP 
spotify_us_df = spotify_us_df[spotify_us_df.isnull().sum()[spotify_us_df.isnull().sum()==0].keys()].copy()

# drop the Country column
spotify_us_df = spotify_us_df.drop(columns=['Country'])

# view the dataframe
spotify_us_df.head()
```

```python
# save it to a csv file
spotify_us_df.to_csv("Resources/spotify_prepared_us_df.csv")
```

```python
# drop title, artist, top10_dummy
spotify_us_df = spotify_us_df.drop(columns = ['Title', 'Artist', 'Top10_dummy'])
spotify_us_df
```

```python
# saw some of the columns needed conversion to float (see Andrew/USA_test-data-prep-work.ipynb)

number_columns = ['Artist_followers', 'Track_number', 'Tracks_in_album', 'danceability', 'energy', 'key', 'loudness' , 'mode', 'speechiness', 'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo', 'duration_ms', 'time_signature']

for column in number_columns:
    spotify_us_df[column] = spotify_us_df[column].astype(float)
```

There is a slight imbalance so will apply random oversampling techniques learned in class

## Prepping the data for models

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# Assign a random_state parameter of 1 to the model
random_oversampler_model = RandomOverSampler(random_state = 1)

# Fit the original training data to the random_oversampler model
X_resampled_train, y_resampled_train = random_oversampler_model.fit_resample(X_train, y_train)
```

Next we scaled the data applying standard scaler techniques

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit(X_resampled_train)
X_train_scaled = X_scaler.transform(X_resampled_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Applying Different Machine Learning Models

SVM Model

```python
from sklearn import svm

# Create the model
model_svm = svm.SVC()

# Fit the model 
model_svm = model_svm.fit(X_train_scaled, y_resampled_train)

# Use the trained model to predict y_pred
y_pred = model_svm.predict(X_test_scaled)
```

```python
from sklearn import metrics 
from sklearn.metrics import classification_report

# print the classification report
print('SVM Model:')
print(classification_report(y_test, y_pred))
```

Decision Tree Classifier Model

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Create a DecisionTreeClassifier model
model_dt = DecisionTreeClassifier()

# Train the model 
model_dt = model_dt.fit(X_train_scaled, y_resampled_train)

# Predict the model 
y_pred = model_dt.predict(X_test_scaled)
```

```python
# Get the metrics/accuracy score
accuracy_dt = metrics.accuracy_score(y_test, y_pred)
print('DecisionTreeClassifier Model:')
print(classification_report(y_test, y_pred))
```

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

# Create a DecisionTreeClassifier model
model_lr = LogisticRegression()

# Train the model 
model_lr = model_lr.fit(X_train_scaled, y_resampled_train)

# Predict the model 
y_pred = model_lr.predict(X_test)
```

```python
# Get the classification report
print('LogisticRegression Model:')
print(classification_report(y_test, y_pred))
```

Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf = model_rf.fit(X_train_scaled,y_resampled_train)
y_pred = model_rf.predict(X_test_scaled)
```

```python
print('RandomForestClassifier Model:')
print(classification_report(y_test, y_pred))
```

K Nearest Neighbors Classifier Model

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
model_knn = model_knn.fit(X_train_scaled,y_resampled_train)
y_pred = model_knn.predict(X_test_scaled)

print('KNeighborsClassifier Model:')
print(classification_report(y_test, y_pred))
```

ADA Boost Classifier Model

```python
from sklearn.ensemble import AdaBoostClassifier

model_adb = AdaBoostClassifier()
model_adb = model_adb.fit(X_train_scaled,y_resampled_train)
y_pred=model_adb.predict(X_test_scaled)

print('AdaBoostClassifier Model:')
print(classification_report(y_test, y_pred))
```

Gradient Boosting Classifier Model

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gb = GradientBoostingClassifier()
model_gb = model_gb.fit(X_train_scaled,y_resampled_train)
y_pred=model_gb.predict(X_test_scaled)

print('GradientBoostingClassifier Model:')
print(classification_report(y_test, y_pred))
```

## Ensemble Model 

```python
from sklearn.ensemble import VotingClassifier

# Ensemble 5 best models together
estimators = [('SVM', model_svm), 
              ('RandomForestClassifier', model_rf), 
              ('AdaBoostClassifier', model_adb), 
              ('DecisionTreeClassifier', model_dt),
              ('GradientBoostingClassifier', model_gb)]
```

```python
# Fit and predict the ensemble model
ensemble_model = VotingClassifier(estimators)
ensemble_model = ensemble_model.fit(X_train_scaled, y_resampled_train)
ensemble_pred = ensemble_model.predict(X_test_scaled)
```

```python
# print the classification report
print(classification_report(y_test, ensemble_pred))
```
