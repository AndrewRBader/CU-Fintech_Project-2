```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', 200)
```

```python
df_us = pd.read_csv(Path('../Resources/spotify_us_df.csv'))
df_us.info()
```

```python
# drop duplicate/redundant columns
df_us_encoded = df_us.drop(columns= ['Title', 'Artist', 'Album/Single', 'Genre', 'Genre_new', 'Explicit', 'Album', 'Release_date', 'Cluster', 'Popu_max', 'Popularity'])
df_us_encoded = df_us_encoded.drop(df_us.columns[0], axis = 1)
df_us_encoded
```

```python
# drop NLP columns created by the creators
df_us_encoded = df_us_encoded[df_us_encoded.isnull().sum()[df_us_encoded.isnull().sum()==0].keys()].copy()
df_us_encoded
```

```python
# Create features (X) and target (y) sets
X = df_us_encoded.drop(columns = ['Top50_dummy'])
y = df_us_encoded['Top50_dummy']
```

```python
# Separate the data
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

## Model 1 - DecisionTreeClassifier

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Create a DecisionTreeClassifier model
model = DecisionTreeClassifier()

# Train the model 
model = model.fit(X_train_scaled, y_train)

# Predict the model 
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn import metrics 
from sklearn.metrics import classification_report
# Get the metrics score
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print(classification_report(y_test, y_pred))
```

## Model 2 - LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
# Create a DecisionTreeClassifier model
model = LogisticRegression()

# Train the model 
model = model.fit(X_train, y_train)

# Predict the model 
y_pred = model.predict(X_test)
```

```python
# Get the accuracy score
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)
# Get the classification report
print(classification_report(y_test, y_pred))
```

## Model 3 - SVM

```python
from sklearn import svm
```

```python
# Create the model
model_svm = svm.SVC()

# Fit the model 
model_svm = model_svm.fit(X_train_scaled, y_train)

# Use the trained model to predict y_pred
y_pred = model_svm.predict(X_test_scaled)
```

```python
# print the classification report
print(metrics.accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
