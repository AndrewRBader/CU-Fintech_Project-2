```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', 200)
```

```python
df_us = pd.read_csv(Path('../Resources/spotify_us_df.csv'))
df_us.info()
```

```python
# drop duplicate/redundant columns
df_us_encoded = df_us.drop(columns= ['Title', 'Artist', 'Album/Single', 'Genre', 'Genre_new', 'Explicit', 'Album', 'Release_date', 'Cluster'])
df_us_encoded = df_us_encoded.drop(df_us.columns[0], axis = 1)
df_us_encoded
```

```python
# drop NLP columns created by the creators
df_us_encoded = df_us_encoded[df_us_encoded.isnull().sum()[df_us_encoded.isnull().sum()==0].keys()].copy()
df_us_encoded
```

```python
# Create features (X) and target (y) sets
X = df_us_encoded.drop(columns = ['Top50_dummy'])
y = df_us_encoded['Top50_dummy']
```

```python
# Separate the data
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Create a DecisionTreeClassifier model
model = DecisionTreeClassifier()

# Train the model 
model = model.fit(X_train, y_train)

# Predict the model 
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics 
from sklearn.metrics import classification_report
# Get the metrics score
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)

print(classification_report(y_test, y_pred))
```
